```python
def find_index_within(lst, key, p):
    """
    This function finds the index of a key in a list, where the difference between the key and the index of the found value is less than or equal to p.

    Parameters:
    lst (list): The list to search.
    key: The value to search for.
    p: The maximum allowed difference between the key and the index of the found value.

    Returns:
    int: The index of the key in the list, or -1 if the key is not found within p indices.
    """
    for i in range(len(lst)-1, -1, -1):
        if abs(key - lst[i]) <= p:
            return i
    return -1
```

**Architecture of my network**

```python
class NeuralNet(nn.Module):

    def __init__(self , Layers , activator = nn.Tanh()):

        super(NeuralNet , self).__init__()

        # Main Neural Network
        # Input Layer
        self.model = nn.ModuleList( [nn.Linear(Layers[0],Layers[1])])

        # Other Layers
        for i in range(1,len(Layers)-1):
            self.model.append(activator)
            self.model.append(nn.Linear(Layers[i],Layers[i+1]))


        # Weight Initializer
        self.model.apply(self.init_weights)

        self.model = self.model.to(device)

    def forward(self , x):  # When Calling The Neural Net
        for layer in self.model:
            x = layer(x)
        return x

    def init_weights(self, m):
        if type(m) == nn.Linear:
            torch.nn.init.xavier_normal_(m.weight , gain = 5/3)
```

**Define function to generate vector Jacobian**

```python
def jacv_gen(x , direc):
    jacv = torch.zeros(x.shape , device = device)
    jacv[:,direc] = 1

    return jacv
```

**Steady State Euler Equation**

```python
from torch.autograd.functional import jvp

def Euler_Equation_1D(model , xcolloc, rho_0, C_0, L, gama, s_func, A_t, epsilon_2):

    global landa, pointwise_Residual

    rho_u_e = model(xcolloc)
    rho = rho_u_e[:,0]
    u = rho_u_e[:,1]
    e = rho_u_e[:,2]
    p = rho * e
    T = T_0 - (u * C_0)**2/(2*Cp)

    # deltaS_R_pred_1 =  np.empty((n_x - 3))
    # deltaS_R_pred_2 =  np.empty((n_x - 3))

    s = s_func(xcolloc)[:,0]
    ds_dx = jvp(s_func, xcolloc, v = jacv_gen(xcolloc , 0) , create_graph=True )[1]
    ds_dx = ds_dx[:,0]

    d_dx = jvp(model, xcolloc, v = jacv_gen(xcolloc,0), create_graph=True)[1]
    Div_u = d_dx[:,1]
    Div_rho = d_dx[:,0]
    Div_p = d_dx[:,0] * d_dx[:,2]
    landa = 1 / (epsilon_2 * (abs(Div_u) - Div_u) + 1)

    continuity_flux = lambda x: (model(x)[:,0]) * s_func(x)[:,0] * (model(x)[:,1])
    dcontinuity_flux_dx = jvp(continuity_flux, xcolloc, v = jacv_gen(xcolloc , 0) , create_graph=True )[1]

    momentum_flux = lambda x: (model(x)[:,0]) * s_func(x)[:,0] * ((model(x)[:,1])**2) + ((model(x)[:,2]) * (model(x)[:,0]) * s_func(x)[:,0]) / gama
    dmomentum_flux_dx = jvp(momentum_flux, xcolloc, v = jacv_gen(xcolloc , 0) , create_graph=True )[1]

    energy_flux = lambda x: (((model(x)[:,0]) * (model(x)[:,2])) / (gama - 1) + (gama * (model(x)[:,0]) * (model(x)[:,1])**2) / 4) * s_func(x)[:,0] * (model(x)[:,1]) + \
                  (model(x)[:,0]) * (model(x)[:,2]) * s_func(x)[:,0] * (model(x)[:,1])
    denergy_flux_dx = jvp(energy_flux, xcolloc, v = jacv_gen(xcolloc , 0) , create_graph=True )[1]

    pointwise_Residual = landa * dcontinuity_flux_dx + landa * (dmomentum_flux_dx - (p * ds_dx) / gama) + landa * denergy_flux_dx

    # for i in range(n_x-3):
    #   deltaS_R_pred_1[i] =  (Cp/R) * math.log(abs(T[i+1] / T[i])) - math.log(abs((rho[i+1] * e[i+1]) / (rho[i] * e[i])))
    #   deltaS_R_pred_2[i] = abs(deltaS_R_pred_1[i]) - deltaS_R_pred_1[i]

    # deltaS_R_pred_1 = torch.tensor(deltaS_R_pred_1,dtype=torch.float32).to(device)
    # deltaS_R_pred_2 = torch.tensor(deltaS_R_pred_2,dtype=torch.float32).to(device)
    # print(deltaS_R_pred_2)
    # print(sum(deltaS_R_pred_2))
    #Flux = lambda x: (continuity_flux, momentum_flux, energy_flux)
    #dflux_dx = jvp(Flux, xcolloc, v = jacv_gen(xcolloc , 0) , create_graph=True )[1]
    #print(dflux_dx)
    #dflux_dx = dflux_dx[:,0]


    return (landa * dcontinuity_flux_dx - 0).pow(2).mean() + (landa * (dmomentum_flux_dx - (p * ds_dx) / gama)).pow(2).mean() + (landa * denergy_flux_dx - 0).pow(2).mean()
```

**Main Loop**

```python
import time
import matplotlib.pyplot as plt

No_epoch = np.array([Adam_it , LBFGS_it]) # Closure (Adams) , L-BFGS

pointwise_Residual_values = [[] for i in range(int((Adam_it + LBFGS_it)/write_step))]
landa_Values = [[] for i in range(int((Adam_it + LBFGS_it)/write_step))]
start = time.time()
total_Loss_values = []
IB_Loss_values = []
PDE_Loss_values = []
epoch_values = []
# y_pred = PINet(X_colloc)

for epoch in range(No_epoch.sum()):

  if epoch < No_epoch[0]:
    optimizer = optimizer1
    scheduler = scheduler1
  else:
    optimizer = optimizer2
    scheduler = scheduler2

  optimizer.step(closure)
  scheduler.step(val_total_Loss)
  total_Loss_values.append(val_total_Loss)
  IB_Loss_values.append(val_IB_Loss)
  PDE_Loss_values.append(val_PDE_Loss)
  epoch_values.append(epoch)
  if(epoch%write_step == 0):
    i = int(epoch/write_step)
    pointwise_Residual_values[i] = pointwise_Residual.detach().numpy()
    landa_Values[i] = landa.detach().numpy()
  if(epoch%1000 == 0):
    elapsed = time.time() - start
    print("Epoch: ", epoch,"total_Loss: ", val_total_Loss, "IB_Loss: ", val_IB_Loss, "PDE_Loss: ", val_PDE_Loss, "Time = ", elapsed)
  if(epoch == (No_epoch.sum()-1)):
    print("Epoch: ", epoch,"Loss: ", val_total_Loss, "Time = ", elapsed)
```
